### RF-distance

```python
import glob
from ete3 import Tree

reference_dir = "/path/to/reference_trees"
result_dirs = glob.glob("/path/to/results/*")

rows = []

for run_dir in result_dirs:
    inferred_tree_file = glob.glob(f"{run_dir}/*tree.newick")[0]
    dataset_name = run_dir.split("/")[-1].split("_")[0]
    reference_tree_file = f"{reference_dir}/{dataset_name}.newick"

    tree_inf = Tree(inferred_tree_file)
    tree_ref = Tree(reference_tree_file)

    rf, max_rf, *_ = tree_inf.robinson_foulds(tree_ref)
    norm_rf = rf / max_rf

    rows.append({
        "dataset": dataset_name,
        "run_dir": run_dir,
        "rf_distance": norm_rf
    })

df_rf = pd.DataFrame(rows)
df_rf
```
